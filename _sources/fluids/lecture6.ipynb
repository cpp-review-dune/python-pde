## While

Conditional statement to execute something while a specific condition is met.

```python
numerical_error = 1.0
epsilon = 1e-2
T = np.zeros(10)
T_new = T.copy() + np.random.random(10)

while numerical_error > epsilon:
    T_new = T_new - T_new / 2
    numerical_error = np.sum(np.abs(T_new - T))
```
